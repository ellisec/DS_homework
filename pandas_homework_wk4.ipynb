```python
import pandas as pd
import numpy as np

#file path
homework_file = "C:/Users/ellis/Downloads/purchase_data.csv"

pymoli_df = pd.read_csv(homework_file)

pymoli_df.head()
```

```python
total_players = len(set(pymoli_df.SN))
total_players
```

```python
# PLayer Count
total_df = pd.DataFrame({"Total_Players":[total_players]})
total_df
```

```python
# Purchasing Analysis (Total)
item_total = len(set(pymoli_df["Item ID"]))
avg_price = pymoli_df.Price.mean()
total_purchase = len(pymoli_df["Item ID"])
total_revenue = pymoli_df.Price.sum()

money_df = pd.DataFrame({"Number_of_Unique_Items":[item_total],"Average_Price":[avg_price],"Number_of_Purchases":[total_purchase],"Total_Revenue":[total_revenue]})

money_df["Average_Price"] = money_df["Average_Price"].map('${:,.2f}'.format)
money_df["Total_Revenue"] = money_df["Total_Revenue"].map('${:,.2f}'.format)

money_df
```

```python
# Gender Demographics
no_dups = pymoli_df.drop_duplicates(["SN"])

males = no_dups.loc[no_dups["Gender"] == "Male", ["SN"]]
male_count = len(males.groupby(["SN"]))

females = no_dups.loc[no_dups["Gender"] == 'Female', ["SN"]]
female_count = len(females.groupby(["SN"]))

other = no_dups.loc[no_dups["Gender"] == "Other / Non-Disclosed", ["SN"]]
other_count = len(other.groupby(["SN"]))

gender_df = pd.DataFrame({"Gender":["Male","Female","Other/Non-Disclosed"],"Total_Count":[male_count,female_count,other_count],"Percentage_of_Players":[(male_count/total_players)*100,(female_count/total_players)*100,(other_count/total_players)*100]})
gender_df = gender_df.set_index(["Gender"])

gender_df
```

```python
# Purchasing Analysis (Gender)

malepurchase = pymoli_df.loc[pymoli_df["Gender"] == "Male", ["Price"]]
malepurchasecount = malepurchase.count()
malepurchasesum = malepurchase.sum()
malepurchaseavg = malepurchase.mean()
malepurchaseper = (malepurchasesum/male_count)

malepurchasecount_value = malepurchasecount.values
malepurchasesum_value = malepurchasesum.values
malepurchaseavg_value = malepurchaseavg.values
malepurchaseper_value = malepurchaseper.values

femalepurchase = pymoli_df.loc[pymoli_df["Gender"] == "Female", ["Price"]]
femalepurchasecount = femalepurchase.count()
femalepurchasesum = femalepurchase.sum()
femalepurchaseavg = femalepurchase.mean()
femalepurchaseper = (femalepurchasesum/female_count)

femalepurchasecount_value = femalepurchasecount.values
femalepurchasesum_value = femalepurchasesum.values
femalepurchaseavg_value = femalepurchaseavg.values
femalepurchaseper_value = femalepurchaseper.values

otherpurchase = pymoli_df.loc[pymoli_df["Gender"] == "Other / Non-Disclosed", ["Price"]]
otherpurchasecount = otherpurchase.count()
otherpurchasesum = otherpurchase.sum()
otherpurchaseavg = otherpurchase.mean()
otherpurchaseper = (otherpurchasesum/other_count)

otherpurchasecount_value = otherpurchasecount.values
otherpurchasesum_value = otherpurchasesum.values
otherpurchaseavg_value = otherpurchaseavg.values
otherpurchaseper_value = otherpurchaseper.values

gender_purchase_df = pd.DataFrame({"Gender":["Male","Female","Other/Non-Disclosed"],"Purchase_Count":[malepurchasecount_value,femalepurchasecount_value,otherpurchasecount_value],"Average_Purchase_Price":[malepurchaseavg_value,femalepurchaseavg_value,otherpurchaseavg_value],"Total_Purchase_Value":[malepurchasesum_value,femalepurchasesum_value,otherpurchasesum_value],"Avg_Total_Purchase_per_Person":[malepurchaseper_value,femalepurchaseper_value,otherpurchaseper_value]})
gender_purchase_df = gender_purchase_df.set_index(["Gender"])

gender_purchase_df
```

```python
#Purchasing Analysis (Age)

bins = [0, 9, 14, 19, 24, 29, 34, 39, 50]
group_labels = ["under 10","10 to 14","15 to 19","20 to 24","25 to 29","30 to 34","35 to 39","over 40"]

pymoli_df["Age_Group"] = pd.cut(pymoli_df["Age"], bins, labels=group_labels)
# pymoli_df.head()

age_sum = pymoli_df.groupby("Age_Group")["Price"].sum()
age_count = pymoli_df.groupby("Age_Group")["Price"].count()
age_avg = pymoli_df.groupby("Age_Group")["Price"].mean()


age_purchase = pd.DataFrame({"Purchase_Count": age_count,"Average_Purchase_Price": age_avg,"Total_Purchase_Value": age_sum})

age_purchase
```

```python
#Top Spenders

player_sum = pymoli_df.groupby("SN")["Price"].sum()
player_count = pymoli_df.groupby("SN")["Price"].count()
player_avg = pymoli_df.groupby("SN")["Price"].mean()

player_purchase = pd.DataFrame({"Total_Count": player_count,"Average_Spend": player_avg,"Total_Spend": player_sum})
#player_purchase.head()

player_purchase_sort = player_purchase.sort_values("Total_Spend", ascending = False)
#player_purchase_sort

player_purchase_sort_top5 = player_purchase_sort.iloc[0:5]
player_purchase_sort_top5
```

```python
#Most Popular Items

item_sum = pymoli_df.groupby("Item ID")["Price"].sum()
item_count = pymoli_df.groupby("Item ID")["Price"].count()
item_cost = pymoli_df.groupby("Item ID")["Price"].mean()
item_name = pymoli_df.groupby("Item ID")["Item Name"].unique()

popular_purchase = pd.DataFrame({"Item_Name": item_name,"Purchase_Count": item_count,"Item_Price": item_cost,"Total_Purchase_Value": item_sum})
# popular_purchase.head()
pop_sort = popular_purchase.sort_values("Purchase_Count", ascending = False)
# pop_sort.head()
pop_sort_top5 = pop_sort.iloc[0:5]
pop_sort_top5
```

```python
# Most Profitable Items

pop_value = popular_purchase.sort_values("Total_Purchase_Value", ascending = False)
# pop_value.head()
pop_value_top5 = pop_value.iloc[0:5]
pop_value_top5
```
